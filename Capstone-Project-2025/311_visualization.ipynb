```python
# === Visualization Notebook: Top Services and Trends ===

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# === Load Data ===
df = pd.read_csv("311_filtered_2023_2025.csv", low_memory=False)
df['requested_date'] = pd.to_datetime(df['requested_date'], errors='coerce')
```

```python
# === Drop Missing Values in Key Columns ===
df = df.dropna(subset=['service_name', 'requested_date', 'comm_name'])
```

```python
# === Focus on Top 20 Service Types ===
top_n = 20
top_services = df['service_name'].value_counts().nlargest(top_n).index
df_top = df[df['service_name'].isin(top_services)].copy()
```

```python
# === Bar Chart: Top 20 Service Types ===
plt.figure(figsize=(12, 6))
df_top['service_name'].value_counts().plot(kind='bar')
plt.title("Top 20 Most Frequent 311 Service Requests")
plt.xlabel("Service Type")
plt.ylabel("Number of Requests")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# === Time Trend: Monthly Requests (Top Services Only) ===
df_top['month'] = df_top['requested_date'].dt.to_period("M")
monthly_trend = df_top.groupby(['month', 'service_name']).size().unstack().fillna(0)

monthly_trend.plot(figsize=(14, 6))
plt.title("Monthly Trend of Top 311 Service Requests")
plt.xlabel("Month")
plt.ylabel("Number of Requests")
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# === Community-wise Breakdown ===
plt.figure(figsize=(14, 6))
df_top['comm_name'].value_counts().head(20).plot(kind='bar')
plt.title("Top 20 Communities by Request Volume")
plt.xlabel("Community")
plt.ylabel("Number of Requests")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
